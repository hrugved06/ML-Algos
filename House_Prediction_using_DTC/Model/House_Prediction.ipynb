```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train
```

```python
train = train.drop('LONGITUDE',axis = 1)
train = train.drop('LATITUDE', axis = 1)
test = test.drop('LONGITUDE',axis = 1)
test = test.drop('LATITUDE', axis = 1)
```

```python
train
```

```python
dummies_train = pd.get_dummies(train['POSTED_BY'],drop_first=True)
dummies_test = pd.get_dummies(test['POSTED_BY'],drop_first=True)
```

```python
train['Owner'] = dummies_train['Owner']
train['Dealer'] = dummies_train['Dealer']
test['Owner'] = dummies_test['Owner']
test['Dealer'] = dummies_test['Dealer']
```

```python
address = [str(train['ADDRESS'][i]) for i in range(len(train))]
address = [str(address[i]).split(',') for i in range(len(train)) ]
address = [address[i][1] for i in range(len(train))]
train['ADDRESS'] = address
```

```python
train = train.drop('POSTED_BY',axis =1)
```

```python
test = test.drop('POSTED_BY',axis =1)
```

```python
address = [str(test['ADDRESS'][i]) for i in range(len(test))]
address = [str(address[i]).split(',') for i in range(len(test)) ]
address = [address[i][1] for i in range(len(test))]
test['ADDRESS'] = address
test
```

```python
dummies_train = pd.get_dummies(train['BHK_OR_RK'])
dummies_test = pd.get_dummies(test['BHK_OR_RK'])
```

```python
train = train.drop('BHK_OR_RK',axis =1)
test = test.drop('BHK_OR_RK',axis =1)
```

```python
train['BHK'] = dummies_train['BHK']
test['BHK'] = dummies_test['BHK']
train['RK'] = dummies_train['RK']
test['RK'] = dummies_test['RK']
```

```python
train
```

```python
test
```

```python
train = train[['UNDER_CONSTRUCTION','RERA','BHK_NO.','SQUARE_FT','READY_TO_MOVE','RESALE','ADDRESS','Owner','Dealer','BHK','RK','TARGET(PRICE_IN_LACS)']]
train
```

```python
test = test[['UNDER_CONSTRUCTION','RERA','BHK_NO.','SQUARE_FT','READY_TO_MOVE','RESALE','Owner','Dealer','BHK','RK']]
test
```

```python
from sklearn.tree import DecisionTreeRegressor
x_train = train[['UNDER_CONSTRUCTION','RERA','BHK_NO.','SQUARE_FT','READY_TO_MOVE','RESALE','Owner','Dealer','BHK','RK']]
y_train = train['TARGET(PRICE_IN_LACS)']
```

```python
reg = DecisionTreeRegressor()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
result = reg.predict(test)
```

```python
submission = pd.read_csv('sample_submission.csv')
```

```python
submission['TARGET(PRICE_IN_LACS)'] = result
```
